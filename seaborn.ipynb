```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/cleaned_all_phones.csv')
df.head(5)
```

```python
df.info()
```

```python
df1=df[["brand","battery","ram(GB)"]]
df=df[df.columns[1:4]]

df.head(5)
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x="brand", data=df)
plt.title("Phone Brands")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20,6))
sns.countplot(x="os", data=df)
plt.title("Phone Brands")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.scatterplot(x="battery", y="ram(GB)", data=df1, color="r", hue="brand", style="brand", s=100)
# sns.regplot(x="battery", y="ram(GB)", data=df1, color="r")
plt.show()
```

```python
plt.figure(figsize=(16,10))
sns.swarmplot(x="brand", y="ram(GB)", data=df1, hue="brand")
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.histplot(x="os", data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
fig, ax = plt.subplots(3,1, figsize=(20,20))
sns.countplot(ax=ax[0], x="os", data=df)
ax[0].tick_params(axis='x', labelrotation=90)
sns.countplot(ax=ax[1], x="brand", data=df)
sns.scatterplot(ax=ax[2], x="battery", y="ram(GB)", data=df1, color="r")
```
